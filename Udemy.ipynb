```python
import pandas as pd
df1 = pd.read_csv("1.csv")
df2 = pd.read_csv("2.csv")
df3 = pd.read_csv("3.csv")
df4 = pd.read_csv("4.csv")
df5 = pd.read_csv("5.csv")
df = pd.concat([df1, df2, df3, df4, df5], ignore_index=True)
```

```python
df.head()
```

```python
df = df.dropna(axis=0, subset=['course_id'])
df.isnull().sum()
```

```python
df.dtypes
```

```python
df['course_id'] = df['course_id'].astype(int)
df['num_subscribers'] = df['num_subscribers'].astype(int)
df['num_reviews'] = df['num_reviews'].astype(int)
df['num_lectures'] = df['num_lectures'].astype(int)
df.dtypes
```

```python
df.drop(columns=['Date'], inplace=True)
```

```python
df['published_timestamp'] = df['published_timestamp'].str.split('T').str[0]
df.dtypes
```

```python
df.isnull().any()
```

```python
df['free/paid'] = df.apply(lambda row: 'free' if row['price'] == 0 else 'paid', axis=1)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df.drop(columns=['Free/Paid'], inplace=True)
df.head()
```

```python
df.to_csv('results.csv', index = False)
```
